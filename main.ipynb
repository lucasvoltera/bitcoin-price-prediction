```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt 

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

from plotly import tools
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
import plotly.graph_objs as go
import gc

import matplotlib.pyplot as plt
import seaborn as sns

#import os
#print(os.listdir("../input"))

from subprocess import check_output
```

```python
import pandas_datareader as web
import datetime as dt
import yfinance as yf

crypto_currency = 'BTC'
against_currency = 'USD'

start = dt.datetime(2016,1,1)
end = dt.datetime.now()

data = yf.download(crypto_currency, start=start, end=end)

# data = web.DataReader(f'{crypto_currency}-{against_currency}','yahoo',start,end)

data.to_csv('data.csv')

df = pd.read_csv('data.csv',parse_dates=[0],
                            index_col='Date')

df.info()
df
```

```python
## Create the VWAP- Volume Weighted Average Price column
df['VWAP'] = (df['Open']*df['Volume']).cumsum() / df['Volume'].cumsum()

## Create the Volume_(Currency) column
df['Volume (Currency)'] = df['Volume'] * df['Close']
df
```

```python
## Check for null values
print(df.isnull().sum())
```

```python
## Fill the null values with the last value available
df['VWAP'].fillna(method='bfill', inplace = True)
df['Volume (Currency)'].fillna(method='bfill', inplace = True)
df
```

```python
# Supondo que 'Timestamp' seja o Ã­ndice do DataFrame
weekly_data = df.resample('W-Mon').mean()

# Se necessÃ¡rio, resete o Ã­ndice para transformar 'Timestamp' de volta em uma coluna
weekly_data.reset_index(inplace=True)
weekly_data.head()
```

```python
# We use Plotly to create the plots https://plot.ly/python/
trace1 = go.Scatter(
    x = weekly_data['Date'],
    y = weekly_data['Open'].astype(float),
    mode = 'lines',
    name = 'Open'
)

trace2 = go.Scatter(
    x = weekly_data['Date'],
    y = weekly_data['Close'].astype(float),
    mode = 'lines',
    name = 'Close'
)
trace3 = go.Scatter(
    x = weekly_data['Date'],
    y = weekly_data['VWAP'].astype(float),
    mode = 'lines',
    name = 'Weighted Avg'
)

layout = dict(
    title='Historical Bitcoin Prices (2015-2018) with the Slider ',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                #change the count to desired amount of months.
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                dict(count=12,
                     label='1y',
                     step='month',
                     stepmode='backward'),
                dict(count=36,
                     label='3y',
                     step='month',
                     stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),
        type='date'
    )
)

data = [trace1, trace2, trace3]
fig = dict(data=data, layout=layout)
iplot(fig, filename = "Time Series with Rangeslider")
```
